# Task 1
Voting Machine
---
Take input from user for age and citizenship and print whether the person is eligible to vote or not.

```python
age = int(input("Enter the age: "))
if age < 18:
  print("You are not eligible to vote")
else:
  citizenship = input("Do you have a citizenship(Yes/No)? ")
  if citizenship.lower() == "yes":
    print("You are eligible to vote")
  else:
    print("You are not eligible to vote")
```

# Task 2
* If age is over 60 and have membership -> 30% discount
* If age is over 60 and have no membership -> 20% discount
* Otherwise "Not eligible for a senior discount"

```python
age = int(input("Enter your age: "))
if age <= 60:
  print("Not Eligible for Senior Citizen Discount")
else:
  membership = input("Do you have a membership(Yes/No)? ")
  if membership.upper() == "YES":
    print("You get 30% Discount")
  else:
    print("You get 20% Discount")
```

# Task 3
Write a Python program to take marks as input from the user and display the grade using if and elif statements according to the following conditions:
* Marks between 80 and 100 -> Grade A
* Marks between 60 and 79 -> Grade B
* Marks between 40 and 59 -> Grade C
* Marks less than 40 -> Grade D

```python
marks = int(input("Enter your marks: "))

if marks < 40:
  print("Grade D")
elif marks <= 59:
  print("Grade C")
elif marks <= 79:
  print("Grade B")
else:
  print("Grade A")
```

# Task 4
* Check if a number is positive and even
* Check if a user is not logged in
* Check if age is between 18 and 60 using and
* Check if a day is Saturday or Sunday using or

```python
num = int(input("Enter a number: "))
if num > 0 and num % 2 == 0:
  print("Positive and Even")
elif num > 0 and num % 2 != 0:
  print("Positive and Odd")
else:
  print("Negative Number")
```

```python
is_logged_in = False
if is_logged_in:
    print("User is logged in")
else:
    print("User is not logged in")
```
