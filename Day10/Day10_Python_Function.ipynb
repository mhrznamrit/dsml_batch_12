  * Takes two number from the user
  * Takes an operator sign (+, -, *, /)
  * Performs the selected operation
  * Displays the result

```python
def operation(a,b,c):
  if c == '+':
    return a+b
  elif c == '-':
    return a-b
  elif c == '*':
    return a*b
  elif c == '/':
    return a/b
  else:
    return "Invalid Operation!"
```

Write a user-defined function that takes 5 user inputs.
  * If the user enters an integer, add it to a list
  * If the user enters a float number, remove the last element

Finally, display the updated list after each operation

```python
def user_list():
  number = []
  for i in range(5):
    value = input(f"Enter value {i+1}: ")
    if "." in value:
      if len(number) > 0:
        removed = number.pop()
        print(f"Float detected, removed: {removed}")
      else:
        print("Float detected, but the list is empty")
    else:
      number.append(int(value))
      print(f"Integer {value} added to the list")
    print("Updated list:", number)
```
