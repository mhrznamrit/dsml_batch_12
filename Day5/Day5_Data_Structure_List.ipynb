# Slicing in List
Slicing is a technique used to extract a specific portion (or
"slice") of a list to create a new list.
---
[start : stop : step]

```python
my_lst = [10,20,30,40,50,60]
```

```python
result1 = my_lst[1:4]
result1
```

```python
result2 = my_lst[:3]
result2
```

```python
result3 = my_lst[3:]
result3
```

```python
result4 = my_lst[0::2]
result4
```

```python
result5 = my_lst[::-1]
result5
```

```python
result6 = my_lst[:]
result6
# Shallow copy
```

# Shallow Copy vs Deep Copy

Screenshot 2026-07-10 161028.png

```python
lst1 = [[1,2],[3,4,5,6]]
lst2 = lst1
print(lst1)
print(lst2)
```

```python
lst3 = [[2,4,6],[1,3,5,7]]
lst4 = lst3.copy()
```

```python
lst4[1][2] = 9
print("Original")
print(lst3)
print("Copy")
print(lst4)
```

```python
import copy
lst5 = [[1,2,3],[4,5,6]]
lst6 = copy.deepcopy(lst5)
lst5[1][2] = 100
```

```python
print ("Original")
print(lst5)
print("Copy")
print(lst6)
```
