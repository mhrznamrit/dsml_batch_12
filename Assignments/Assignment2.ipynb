# Student Result Management System
Create a user-defined function called student_result() that:
1. Takes the names and marks of 5 students as input.
2. Stores student data in a dictionary.
3. Uses conditional statements to assign grades:
    * A:80 and above
    * B:60–79
    * C:40–59
    * F:Below 40
4. Displays each student's name, marks, and grade.
5. Calculates and displays:
    * Highest marks
    * Lowest marks
    * Average marks

```python
def student_result():
  student_data = {}

  print("Enter Student Details:")
  for i in range(5):
    name = input(f"Enter name of student {i+1}: ")

    while True:
      marks = int(input(f"Enter marks for {name}: "))
      if marks >= 0 and marks <= 100:
        break
      else:
        print("Please enter valid mark between 0 and 100")
    student_data[name] = marks

  print("\nStudent's Results:")
  for name, marks in student_data.items():
    if marks >= 80:
      grade = 'A'
    elif marks >= 60:
      grade = 'B'
    elif marks >= 40:
      grade = 'C'
    else:
      grade = 'F'
    print(f"Name: {name} | Marks: {marks} | Grade: {grade}")

  print("\nClass Statistics:")
  scores = list(student_data.values())
  highest_marks = max(scores)
  lowest_marks = min(scores)
  average_marks = sum(scores)/len(scores)
  print(f"Highest Marks: {highest_marks}")
  print(f"Lowest Marks: {lowest_marks}")
  print(f"Average Marks: {average_marks}")

student_result()
```

# Smart Shopping Cart System
Create a user-defined function called shopping_cart() that:
1. Allows the user to enter 5 product prices.
2. Stores the prices in a list.
3. Uses a loop to calculate the total bill.
4. Applies discounts using conditional statements:
    * Total ≥ 5000 → 15% discount
    * Total ≥ 3000 → 10% discount
    * Total ≥ 1000 → 5% discount
    * Otherwise → No discount
5. Displays:
    * Product list
    * Total amount
    * Discount amount
    * Final payable amount

```python
def shopping_cart():
  price = []

  print("Enter Product Prices:")
  for i in range(5):
    while True:
      p = float(input(f"Enter price of product {i+1}: "))
      if p >= 0:
        price.append(p)
        break
      else:
        print("Price cannot be negative. Try again.")

  total_amount = sum(price)

  if total_amount >= 5000:
    discount = 0.15
  elif total_amount >= 3000:
    discount = 0.10
  elif total_amount >= 1000:
    discount = 0.05
  else:
    discount = 0.0

  discount_amount = total_amount * discount
  final_amount = total_amount - discount_amount

  print("\nCheckout Summary:")
  print(f"Product List: {price}")
  print(f"Total amount: {total_amount}")
  print(f"Discount amount: {discount_amount}")
  print(f"Final payable amount: {final_amount}")

shopping_cart()
```
