# CONDITIONAL STATEMENT

```python
# if
# elif
# else
```

```python
number = 10

if (number>20):
  print("here")
else:
  print("there")
```

```python
age = 20
if (age>=18):
  print("Eligible to vote")
```

```python
age = 25
if age <= 12:
  print("Child")
elif age <= 19:
  print("Teenager")
elif age <= 35:
  print("Young Adult")
elif age <= 65:
  print ("Adult")
else:
  print ("Elderly")
```

# Logical Operation

```python
# AND Operator

age = 20
citizenship = True

if age >= 18 and citizenship:
    print("You are eligible to vote")
else:
    print("You are not eligible to vote")
```

```python
# OR Operator

username = "admin"
password = "1234"

if username == "admin" or password == "1234":
    print("Login successful")
else:
    print("Invalid credentials")
```

```python
# NOT Operator

is_logged_in = False
if not is_logged_in:
    print("Please Log In")
else:
    print("Welcome")
```

**any() function with if**

```python
t = {1,2,3,4,5}
if any(a>3 for a in t):
  print("Good")
```

**all() function with if**

```python
marks = [40,70,60,90]
if all(mark>=40 for mark in marks):
  print("All students passed")
```
