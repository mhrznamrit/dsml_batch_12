# Write a function that takes three numbers as arguments and returns the largest number

```python
def large(a,b,c):
  if a > b and a > c:
    largest = a
  elif b > c:
    largest = b
  else:
    largest = c
  print(f"The largest number is {largest}")

large(1748843,1737483,1743839)
```

# Write a function that takes a string and returns the number of vowels (a, e, i, o, u) present in it

```python
def count_vowels(word):
  vowels = "aeiouAEIOU"
  count = 0
  for char in word:
    if char in vowels:
      count += 1
  return count

count_vowels("HIppopotaMUS")
```
