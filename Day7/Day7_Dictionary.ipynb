# Joining two dictionaries
* Using update() method
* Using **kwargs to unpack

```python
dict1 = {
    "name" : "ram",
    "address" : "ktm",
    "age" : 25
}
dict1
```

```python
dict2 = {
    "phone" : 9860,
    "skill" : "python"
}
dict2
```

```python
dict1.update(dict2)
```

```python
dict1
```

```python
dict3 = {
    "name" : "shyam",
    "phone" : 9700
}
dict3
```

```python
dict3.update(dict1)
```

```python
dict3
```

```python
dict4 = {
    "gender" : "male",
    "marks" : 90
}
dict4
```

```python
student_dict = {**dict1, **dict2, **dict3, **dict4}     # using kwargs to unpack
student_dict
```

```python
d1 = {
    "name" : "ram",
    "age" : 20
}
d2 = {
    "name" : "shyam",
    "class" : 12
}
d3 = {
    "address" : "ktm",
    "phone" : 9860
}
```

```python
dict_student = {**d1, **d2, **d3}
dict_student
```

```python
student = {"z" : 20, "p" : 80, "a" : 60}
result = sorted(student)
result
```

```python
# merge dictionaries
d1 | d2
```
