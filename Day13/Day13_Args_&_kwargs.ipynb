# *args and **kwargs

```python
def sum_total(*num):
  result = 0
  for n in num:
    result += n
  return result
```

```python
sum_total(1,2,3,4,5,6)
```

```python
def add(*args):
  total = sum(args)
  return total

add(1,3,7)
```

```python
def func(**kwargs):
  for k, val in kwargs.items():
    print(k, "=", val)
```

```python
func(s1='Python', s2='is', s3='Awesome')
```

```python
def introduce(**kwargs):
  details = []
  for k, v in kwargs.items():
    details.append(k + ":" + str(v))
  return " | ".join(details)
```

```python
introduce(Name="Alex", Age="25", City="London")
```

```python
def student_data(*subject, **student_details):
  print("Subjects:")
  for s in subject:
    print(s)
  print("\nStudent Information:")
  for k, v in student_details.items():
    print(k, ":", v)
```

```python
student_data("Math", "Science", "Nepali", Name="Leo", Class="VIII", Phone="9876543210")
```

* Write a function that accepts products and their prices as keyword arguments and returns the total.

```python
def products(**kwargs):
  price = 0.0
  for k, v in kwargs.items():
    print(k, ":", float(v))
    price += v
  print(f"The total price is {price}")
```

```python
products(Apple=500, Potato=180, Biscuit=50, Noodles=120)
```

* Write a function cart_total(*args) that accepts the prices of any number of products and returns the total bill amount.

```python
def cart_total(*args):
  amount = sum(args)
  print(f"The total bill amount is {amount}")

headphone = 1500
bottle = 800
charger = 1050
shoes = 2499
locket = 999

cart_total(headphone, bottle, charger, shoes, locket)
```

* Create a function product_info(**kwargs) that accepts product details such as name, price, brand, and category, then displays them in a readable format.
