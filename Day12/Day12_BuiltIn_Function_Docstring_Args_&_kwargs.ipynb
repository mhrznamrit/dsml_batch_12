# BuiltIn Functions

```python
# enumerate
# zip
```

```python
data1 = ["ram", "shyam", "gita", "rita"]
data2 = [20, 30, 40, 50]

for i in zip(data1, data2):
  print(i)
```

```python
for i, (name,age) in enumerate(zip(data1, data2), start = 1):
  print(i, name, age)
```

```python
names = ["John", "Seth", "Roman", "Dean", "Kelly"]
age = [49, 42, 45, 39, 52]
gender = ["Male", "Male", "Male", "Male", "Female"]
salary = [50000, 80000, 100000, 60000, 35000]

for i, (name,age,gender,salary) in enumerate(zip(names,age,gender,salary), start = 1):
  print(f"{i}. Name: {name} | Age: {age} | Gender: {gender} | Salary: {salary}")
```

```python
# map() : map(function, iterable)

def mux(x):
  return x*x

number = [1,2,3,4,5]
map_object = map(mux, number)
map_object
```

```python
print(list(map_object))
```

```python
# filter()

def odd(x):
  return x % 2 != 0

sample = (1,9,10,17,20,24,28,32,44,49)
odd_num = filter(odd, sample)
print(list(odd_num))
```

```python
# reduce()

from functools import reduce
mylist = [5,6,7,8,9]
def add(x,y):
  return x+y

added = reduce(add, mylist)
added
```

```python
# Classwork

from functools import reduce
def maxi(a,b):
  if a > b:
    return a
  else:
    return b

numbers = [4, 9, 2, 8, 1]
maximum_number = reduce(maxi, numbers)
maximum_number
```

# Doc String

```python
# doc string

def calculate_area(length, breadth):
  """
  Calculate the area of the rectangle:

  Parameters:
  length(float): Length of the rectangle
  breadth(float): Breadth of the rectangle

  return:
  float: Area of the rectangle
  """
  return length * breadth
```

```python
print(calculate_area.__doc__)
```

```python
help(calculate_area)
```

# *args

```python
def display(*data):
  print(data)
```

```python
display("Amrit", "Maharjan", 27)
```

```python
def area(*sides):
  if len(sides) == 2:
    l,w = sides
    print("2D Area = ", l*w)
  elif len(sides) == 3:
    l,w,h = sides
    print("3D Volume = ", l*w*h)
  else:
    print("Invalid number of argument")
```

```python
area(3,4)
```

```python
area(5,6,7)
```

```python
def mux(*args):
  result = 1
  for num in args:
    result *= num
  return result
```

```python
mux(1,2,3,4,5)
```

```python
mux(3,4,5,6,7,8,9,10)
```
