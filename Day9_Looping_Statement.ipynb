# While Loop

```python
num = 20

while num > 17:
  print ("hello")
  num = num - 1
```

```python
n = 0
while (n < 10):
  n = n + 1
  print(n)
```

```python
count = 1
while count < 10:
  if count == 5:
    break
  print(count)
  count += 1
print ("Loop End")
```

```python
count = 0
while count < 10:
  count += 1
  if count == 3:
    continue
  print(count)
print ("Loop End")
```

# Number Guessing Game
* Create a program where:
  * A secret number is stored inside the program.
  * User keeps entering numbers.
  * If the number matches, print: "Correct Guess!"
  * Use break to stop the loop
  * If wrong, print: "Try Again"

```python
secret_number = 47

while True:
  num = int(input("Enter a number: "))

  if secret_number == num:
    print("Correct Guess!")
    break
  else:
    print("Try Again")
```

# ATM Menu System
Create a simple ATM menu:
1. Check Balance
2. Deposit
3. Exit

```python
balance = 2000

while True:
  print("ATM Menu\n1. Check Balance \n2. Deposit \n3. Exit")
  entry = int(input("Enter your choice "))

  if entry == 1:
    print("Your balance is:")
    print(balance)
    print("\n")
  elif entry == 2:
    deposit = int(input("Enter deposit amount: "))
    balance = balance + deposit
    print("Amount deposited! \nYour new balance is:")
    print(balance)
    print("\n")
  elif entry == 3:
    print("Thank you for using our service")
    break
  else:
    print("Invalid choice")
    print("\n")
```

# For-Range Loop

```python
for i in range(10):
  print(i)
```

```python
for i in range(1,11):  # (start, stop)
  print(2, "*", i, "=", i*2)
```

```python
for i in range(2,11,2): # (start, stop, step)
  print(i)
```

```python
for i in range(10, 0, -1):
  print (i)
```
