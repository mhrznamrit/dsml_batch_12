```python
# enumerate()

names = ["Aron", "Ben","Chris"]
list(enumerate(names))
for (i, name) in enumerate(names):
  print(f"{name} : Roll {i+1}")
```

```python
# zip()

data1 = [1,2,3,4,5]
data2 = ["A","B","C","D","E"]

for (dat,num) in zip(data1,data2):
  print(dat, num, sep =" --> ")
```

```python
# map()

numbers = [2, 4, 6, 8, 10]
def mult(x):
  return x ** 2
map_obj = map(mult, numbers)
print(list(map_obj))
```

```python
# filter()

def even(num):
  return num % 2 == 0
f_obj = filter(even, [1,3,4,6,8,9])
list(f_obj)
```

```python
# reduce()

from functools import reduce
mylist = [1,2,3,4,5]

def factorial(x,y):
  return x * y

red_obj = reduce(factorial, mylist)
red_obj
```

```python
# Find Maximum

from functools import reduce

def maximum(x, y):
  if x > y:
    return x
  return y

print(reduce(maximum, [4, 7, 2, 8, 1, 8, 9]))
```
